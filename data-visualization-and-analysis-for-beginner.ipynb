# Data Analysis

```python
print(data.info()) # find Nan, number of columns object(7)
print(data.describe()) # check the values
pd.set_option('display.max_columns', 8) # display all columns (object +1 = 8)
print(data.head())
```

# Data Visualization

```python
# 'Age' is numerical variable types so it uses Histogram or Boxplot are better
plt.hist(data['Age'], 20) # 20 is for bins (number of vertical lines)
plt.axvline(data['Age'].mean(), color='k', linestyle="dashed") # for check the mean value and standardization
plt.xlim(xmin=5, xmax=90) # column range 5 to 90
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Number of Age')
plt.show()
```

```python
# This graph is right-skewed and around 30 is the mean value. We can guess between 60 and 80 are outliers but not sure.
# We need more careful analysis.
```

```python
# One Categorical value type use Bar(Ordinal) or Pareto chart (Nominal) are better 
category = data.drop(columns=['Timestamp', 'Participant_ID', 'Age']) # remove data
for itr in category:
    fig, (ax1, ax2) = plt.subplots(1, 2) 
    data[itr].value_counts().plot(kind="barh", ax=ax1)

# One Numerical and One Categorical value uses Histograms or Boxplots are better
    data.boxplot(column=['Age'], by=[itr], ax=ax2)
    if itr == 'Nationality':
        plt.xticks(rotation=90)
    fig.suptitle('Bar and Boxplot By ' + itr)
    plt.show()
    
# Two Categorical uses Cluster bar or Stacked bar chart are better
    if itr != 'Gender':
        category.groupby(['Gender', itr]).size().unstack().plot(kind='bar', stacked=True)
        plt.show()
```
